```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn

data = pd.read_csv('DOptExperimnentResults.csv',index_col = 0)
data
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tqdm import tqdm 

#Normalizing data
scaler = MinMaxScaler()

x =  scaler.fit_transform(data[['K','percentDMSO','Vol']])
y = scaler.fit_transform(data[['r_squared']])

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)

X_train = torch.tensor(X_train, dtype=torch.float)
y_train = torch.tensor(y_train, dtype=torch.float)

X_test = torch.tensor(X_test, dtype=torch.float)
y_test = torch.tensor(y_test, dtype=torch.float)

Lin = nn.Linear(3,1)
opt = torch.optim.SGD(Lin.parameters(),lr=1e-3)
loss_fn = nn.MSELoss()


for i in tqdm(range(10000)):
    y_pred = Lin.forward(X_train)
    loss = loss_fn(y_train,y_pred)
    loss.backward()
    opt.step()
    opt.zero_grad()

    

    
print('Training loss = ' ,loss.item())
print()
y_pred = Lin.forward(X_test)
loss = loss_fn(y_test,y_pred)
plt.figure(figsize=(5,5))
plt.scatter(y_test,y_pred.detach())
plt.plot((0,1),(0,1))
plt.xlabel('actual')
plt.ylabel('predicted')
plt.ylim(0,1)
plt.xlim(0,1)
plt.show()

print('Test Loss = ',loss.item())

print('\nWeights: ')
print(Lin.weight)
print('\nBias: ')
print(Lin.bias)

pd.DataFrame(Lin.weight.detach().numpy(), columns = ['K','percentDMSO','Vol'])
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2)

X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.fit_transform(X_test)

X_train_poly = torch.tensor(X_train_poly,dtype=torch.float)
X_test_poly = torch.tensor(X_test_poly,dtype=torch.float)

Non_Lin = nn.Linear(10,1)
opt = torch.optim.Adam(Lin.parameters(),lr=1e-3)


for i in range(10_000):
    y_pred = Non_Lin.forward(X_train_poly)
    loss = loss_fn(y_train,y_pred)
    print(loss.item(),end='\r')
    loss.backward()
    opt.step()
    opt.zero_grad()

    

    
print('Training loss = ' ,loss.item())
print()
y_pred = Non_Lin.forward(X_test_poly)
loss = loss_fn(y_test,y_pred)
print('Test Loss = ',loss.item())

print('\nWeights: ')
print(Non_Lin.weight)
print('\nBias: ')
print(Non_Lin.bias)
```
