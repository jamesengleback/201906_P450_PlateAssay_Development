```python
import pandas as pd

data = pd.read_csv('EchoLog5thNov.txt',delimiter = '\t',error_bad_lines=False,header=None)
data.columns = ['Time','Program','Log type','Message']


def GetSkippedWells(data):
    WarningMessages = data[data['Log type'] == 'WARN']['Message']
    Skipped = WarningMessages.loc[WarningMessages.str.contains('skipping')]
    return Skipped
    
GetSkippedWells(data)
```
