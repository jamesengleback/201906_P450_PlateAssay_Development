```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


Titrations = pd.read_csv('TitrationMetrics.csv')

Plates_1 = pd.read_csv('Triton0.1Percent_metrics.csv',index_col = 0)
Plates_2 = pd.read_csv('Triton1Percent_metrics.csv',index_col = 0)

Plates_1['Percent Triton'] = 0.1
Plates_1.loc[Plates_1['Plate'] == '20191202_KPi.CSV','Percent Triton'] = 0
Plates_2['Percent Triton'] = 1


Titrations.loc[Titrations['Unnamed: 0'].str.contains('4PHENYLIMIDAZOLE'),'Compound'] = '4-Phenylimidazole'
Titrations.loc[Titrations['Unnamed: 0'].str.contains('Lauric'),'Compound'] = 'Lauric Acid'
Titrations.loc[Titrations['Unnamed: 0'].str.contains('ArachadonicaAcid'),'Compound'] = 'Arachadionic Acid'
Titrations.loc[Titrations['Unnamed: 0'].str.contains('PalmiticAcid'),'Compound'] = 'Palmitic Acid'
Titrations.drop('Unnamed: 0',inplace = True, axis = 1)

Plates = Plates_2.append(Plates_1).loc[:,['Compound','Kd','Noise','R_sq','Vmax','Percent Triton']]


for i in ['4-Phenylimidazole','Lauric Acid','Arachadionic Acid','Palmitic Acid']:
    temp_titrations = Titrations.loc[Titrations['Compound'] == i]
    temp_plate1 = Plates.loc[Plates['Compound'] == i]
    temp_plate1 = temp_plate1.loc[temp_plate1['Percent Triton'] == 0]
    
    temp_plate2 = Plates.loc[Plates['Compound'] == i]
    temp_plate2 = temp_plate2.loc[temp_plate2['Percent Triton'] == 0.1]
    
    temp_plate3 = Plates.loc[Plates['Compound'] == i]
    temp_plate3 = temp_plate3.loc[temp_plate3['Percent Triton'] == 1.0]
    
    plt.figure(figsize = (7,4))
    plt.hist(temp_titrations.loc[:,'Km'],alpha = 0.5)
    plt.hist(temp_plate1.loc[:,'Kd'],alpha = 0.5)
    plt.hist(temp_plate2.loc[:,'Kd'],alpha = 0.5)
    plt.hist(temp_plate3.loc[:,'Kd'],alpha = 0.5)
    plt.title(i)
    plt.xlabel('Kd')
    plt.ylabel('Frequency')
    plt.legend(['Titrations','Plate 0 % Triton','Plate 0.1 % Triton','Plate 1 % Triton'])
    plt.show()
```
