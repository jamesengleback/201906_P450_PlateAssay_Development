```python
import os
root_path = '../1_PlateSelection/'
file_paths = sorted([os.path.join(root_path, i) for i in os.listdir(root_path)])
file_paths
```

```python
from mimetypes import guess_type

csv_paths = [i for i in file_paths if guess_type(i)[0] == 'text/csv']
csv_paths
```

```python
spec_data_file_paths = list(filter(lambda x : 'BM3' in x, csv_paths))
spec_data_file_paths
```

```python
import pandas as pd
df = pd.read_csv(spec_data_file_paths[0])
df.head()
```

```python
def parse_spec(path: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    initial_columns = df.columns 
    wavelength_col_num = df.iloc[0,:].to_list().index('Wavelength (nm)')
    wavelengths_ = df.iloc[:, wavelength_col_num]
    wavelengths = wavelengths_.loc[wavelengths_.str.match('[0-9\.]+')]
    df = df.loc[wavelengths.index, :]
    df = df.loc[:, 1::2]
    return df

parse_spec(spec_data_file_paths[0])
```
